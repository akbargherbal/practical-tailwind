```python
list_index = []
for root, dirs, files in os.walk('./pages/'):
    for file in files:
        if file.endswith('.html') and (Path(file).name.startswith('00_index')):
            list_index.append(os.path.join(root, file).replace(os.sep, '/'))

from bs4 import BeautifulSoup
from typing import Optional, Dict, Callable
import os
def insert_html_tag(
    file_path: str,
    tag_html: str,
    into: str = 'head',
    selector: Optional[str] = None,
    prepend: bool = False
) -> None:
    """
    Inserts `tag_html` into a section of the HTML.

    - `into`: a tag like 'head', 'body', 'div', etc.
    - `selector`: an optional CSS-style selector (e.g., '#main', '.container') to find the exact tag to insert into.
    - `prepend`: if True, insert at beginning; else append at end.
    """
    with open(file_path, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    # Fallback: create <html> if it doesn't exist
    if not soup.html:
        html_tag = soup.new_tag("html")
        html_tag.contents = soup.contents
        soup.contents = [html_tag]

    # Choose the section
    target = None
    if selector:
        target = soup.select_one(selector)
    if not target:
        target = soup.find(into)
    if not target:
        target = soup.new_tag(into)
        soup.html.append(target)

    # Add the new tag
    new_tag = BeautifulSoup(tag_html, 'html.parser')
    if prepend:
        target.insert(0, new_tag)
    else:
        target.append(new_tag)

    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(str(soup.prettify()))



def remove_html_tag(file_path: str, tag_name: str, attrs: Optional[Dict] = None, match_func: Optional[Callable] = None) -> None:
    """
    Removes tags matching `tag_name`, optionally filtering by attributes or a custom `match_func`.
    
    - `attrs`: e.g., {'id': 'main', 'class': 'sidebar'}
    - `match_func`: optional function(tag) -> bool to apply extra filtering (e.g., text match)
    """
    with open(file_path, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')

    tags = soup.find_all(tag_name, attrs=attrs or {})

    for tag in tags:
        if match_func is None or match_func(tag):
            tag.decompose()

    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(str(soup.prettify()))



def add_class_to_elment(path, element, class_name):
    '''
    Add a new class to a given html element;
    '''
    with open(path, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f, 'html.parser')
    try:
        soup.find(element).attrs['class'].append(class_name)
    except:
        soup.find(element).attrs['class'] = [class_name]
    
    with open(path, 'w', encoding='utf-8') as f:
        f.write(str(soup.prettify()))
    

    

```
